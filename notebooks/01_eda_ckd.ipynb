```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the datasets
# You have the main file and the train/test splits
df_main = pd.read_csv('/content/drive/MyDrive/Dataset/Kidney dataset/Chronic_Kidney_Dsease_data.csv')
df_train = pd.read_csv('/content/drive/MyDrive/Dataset/Kidney dataset/kidney_disease_train.csv', header=None) # The split files appear to lack headers

print("--- 1. SHAPE & TYPES ---")
print(f"Main Dataset Shape: {df_main.shape}")
print(df_main.info())
print("\n")

print("--- 2. MISSING VALUES ---")
missing_values = df_main.isnull().sum()
print("Columns with missing data:")
print(missing_values[missing_values > 0])
print("\n")

print("--- 3. TARGET DISTRIBUTION ---")
# In the main dataset, the target appears to be 'Diagnosis'
print("Main Dataset 'Diagnosis' distribution:")
print(df_main['Diagnosis'].value_counts(normalize=True))

# In the train dataset, the target is the last column ('ckd' vs 'notckd')
print("\nTrain Dataset target distribution:")
print(df_train.iloc[:, -1].value_counts(normalize=True))
print("\n")

print("--- 4. BASIC STATISTICS ---")
display(df_main.describe())

# 2. Start building the Markdown string
md_content = "# Data Quality & Known Issues Report\n\n"
md_content += "This document tracks data issues discovered during the Exploratory Data Analysis (EDA) phase.\n\n"

# --- MISSING VALUES ---
md_content += "## 1. Missing Values\n"
md_content += "The datasets contain missing values that will need to be handled via imputation or dropping before model training.\n\n"

md_content += "### Main Dataset (`Chronic_Kidney_Dsease_data.csv`)\n"
missing_main = df_main.isnull().sum()
missing_main = missing_main[missing_main > 0].sort_values(ascending=False)
if not missing_main.empty:
    md_content += missing_main.to_frame(name="Missing Count").to_markdown() + "\n\n"
else:
    md_content += "No missing values detected in the main dataset.\n\n"

md_content += "### Training Dataset (`kidney_disease_train.csv`)\n"
missing_train = df_train.isnull().sum()
missing_train = missing_train[missing_train > 0].sort_values(ascending=False)
if not missing_train.empty:
    md_content += missing_train.to_frame(name="Missing Count").to_markdown() + "\n\n"
else:
    md_content += "No missing values detected in the training dataset.\n\n"

# --- TARGET DISTRIBUTION ---
md_content += "## 2. Target Distribution (Class Imbalance)\n"
md_content += "Monitoring the distribution of the target variable to decide if techniques like SMOTE or class weighting are necessary.\n\n"

md_content += "### Main Dataset Target (`Diagnosis`)\n"
target_main = (df_main['Diagnosis'].value_counts(normalize=True) * 100).round(2)
md_content += target_main.to_frame(name="Percentage (%)").to_markdown() + "\n\n"

md_content += "### Training Dataset Target (`classification`)\n"
target_train = (df_train[25].value_counts(normalize=True) * 100).round(2)
md_content += target_train.to_frame(name="Percentage (%)").to_markdown() + "\n\n"

# --- OUTLIERS & DATA TYPES ---
md_content += "## 3. Data Type & Formatting Issues\n"
md_content += "* Note: Some columns in the training dataset may contain string representations of missing values (e.g., `?` or `\t?`) or incorrect types that will need converting to numeric.\n"
md_content += "* Patient IDs should be dropped prior to training to prevent data leakage.\n"

# 3. Write to known-issues.md
with open('known-issues.md', 'w') as f:
    f.write(md_content)
```
